```python
import pandas as pd
from pathlib import Path
```

```python
directory = '../Datos'
dir_files = Path(directory).glob('*.csv')
files_ok = []
files_na = []
for f in dir_files:
    file = pd.read_csv(f)
    print(f)
    print(file.info())
    print(file.isna().sum())
    print('==================================================')
    if file.isna().sum().sum():
        files_na.append(file)
    else:
        files_ok.append(file)
```
